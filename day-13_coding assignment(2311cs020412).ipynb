```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the Dataset
file_path = 'Day_13_Pharma_data.csv'  # Update this to your file's location
pharma_data = pd.read_csv(file_path)

# Step 2: Inspect and Clean the Data
# Display basic information
print(pharma_data.info())
print(pharma_data.head())

# Check for missing values
print("\nMissing Values:\n", pharma_data.isnull().sum())

# Drop duplicates if present
pharma_data = pharma_data.drop_duplicates()

# Step 3: Visualizations

# 1. Bar Plot: Total Sales per Region
region_sales = pharma_data.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x=region_sales.index, y=region_sales.values, palette="viridis")
plt.title('Total Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# 2. Scatter Plot: Marketing Spend vs. Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pharma_data, x='Marketing_Spend', y='Sales', hue='Region', palette='cool')
plt.title('Marketing Spend vs Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.show()

# 3. Boxplot: Drug Effectiveness by Age Group
plt.figure(figsize=(8, 6))
sns.boxplot(data=pharma_data, x='Age_Group', y='Effectiveness', palette='Set2')
plt.title('Drug Effectiveness by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Effectiveness')
plt.show()

# 4. Line Plot: Sales Trend for Each Product over Trial Periods
plt.figure(figsize=(10, 6))
sns.lineplot(data=pharma_data, x='Trial_Period', y='Sales', hue='Product', marker='o', palette='tab10')
plt.title('Sales Trends Over Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Sales')
plt.show()

# 5. Heatmap: Correlation Between Sales, Marketing Spend, and Effectiveness
correlation_data = pharma_data[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Step 4: Summarize Observations
print("\nObservations:")
print("1. Bar Plot: Total sales are distributed differently across regions.")
print("2. Scatter Plot: There is a visible relationship between Marketing Spend and Sales.")
print("3. Boxplot: Drug effectiveness varies across different age groups.")
print("4. Line Plot: Sales trends show variation over trial periods for each product.")
print("5. Heatmap: Correlation matrix reveals relationships among Sales, Marketing Spend, and Effectiveness.")
```
