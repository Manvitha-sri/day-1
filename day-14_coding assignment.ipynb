```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Day_14_Pharma_data.csv'  # Replace with the path to your file
data = pd.read_csv(file_path)

# Ensure numeric columns are properly converted
data['Effectiveness'] = pd.to_numeric(data['Effectiveness'], errors='coerce')
data['Side_Effects'] = pd.to_numeric(data['Side_Effects'], errors='coerce')
data['Marketing_Spend'] = pd.to_numeric(data['Marketing_Spend'], errors='coerce')

# Setting up the visualization style
sns.set_theme(style="whitegrid")

# 1. Bar plot comparing average Effectiveness for each drug across different regions
avg_effectiveness = data.groupby(['Product_Name', 'Region'])['Effectiveness'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_effectiveness, x='Product_Name', y='Effectiveness', hue='Region')
plt.title('Average Effectiveness of Drugs by Region', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Average Effectiveness', fontsize=12)
plt.legend(title='Region', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Violin plot to show distribution of Effectiveness and Side Effects for each product
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Product_Name', y='Effectiveness', inner='box', palette='muted')
plt.title('Distribution of Effectiveness by Product', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Effectiveness', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Product_Name', y='Side_Effects', inner='box', palette='muted')
plt.title('Distribution of Side Effects by Product', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Side Effects', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(data, vars=['Effectiveness', 'Side_Effects', 'Marketing_Spend'], hue='Product_Name', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()

# 4. Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Trial_Period', y='Effectiveness', palette='coolwarm')
plt.title('Effectiveness by Trial Period', fontsize=14)
plt.xlabel('Trial Period', fontsize=12)
plt.ylabel('Effectiveness', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='Marketing_Spend', y='Effectiveness', scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Effect of Marketing Spend on Effectiveness', fontsize=14)
plt.xlabel('Marketing Spend', fontsize=12)
plt.ylabel('Effectiveness', fontsize=12)
plt.tight_layout()
plt.show()
```
